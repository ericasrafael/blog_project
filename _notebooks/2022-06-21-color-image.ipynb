```python
#hide_input

def grayscale(image_colored):
    # pegando as dimenses (largura, altura)
    w, h = image_colored.size
    # criando nova imagem para manter a original
    img = Image.new("RGB", (w,h))
    # capturando pixels
    for x in range(w):
        for y in range(h):
            # capturando cada pixel percorrido
            pxl = image_colored.getpixel((x,y))
            # mdia das coordenadas RGB de cada pixel
            lum = ((pxl[0]+pxl[1]+pxl[2])//3)
            #  Modificando o pixel em x, y
            img.putpixel((x,y), (lum,lum,lum))
    return img
```

```python
#hide_input

def grayscale_ponderada(image_colored):
    # pegando as dimenses (largura, altura)
    w, h = image_colored.size
    # criando nova imagem para manter a original
    img = Image.new("RGB", (w,h))
    # capturando pixels
    for x in range(w):
        for y in range(h):
            # capturando cada pixel percorrido
            pxl = image_colored.getpixel((x,y))
            # mdia das coordenadas RGB de cada pixel
            lum = int(0.3*pxl[0]+0.59*pxl[1]+0.11*pxl[2])
            #  Modificando o pixel em x, y
            img.putpixel((x,y), (lum,lum,lum))
    return img
```

```python
#hide_input

grayscale(colorida_menor)
```

```python
#hide_input

grayscale_ponderada(colorida_menor)
```
